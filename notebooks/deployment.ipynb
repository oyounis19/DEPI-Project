```python
import os
from threading import Thread
from pyngrok import ngrok
```

```python
ngrok.set_auth_token('api_key')
```

```python
def run_streamlit():
    os.system('streamlit run /kaggle/working/DEPI-Project/app.py --server.port 8501')
```

```python
# Run app.py in the background
thread = Thread(target=run_streamlit)
thread.start()
```

```python
# Create a tunnel to view the webpage publicly
public_url = ngrok.connect(addr='8501', proto='http')
print('App is live at: ' + str(public_url))
```
